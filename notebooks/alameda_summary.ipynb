# Address detection: Alameda

I created a 1% sample of row-wise and column-wise extracted data.  This allows a representative look at rows from the cleaned data set in order to determine an appropriate regex that would match most addresses in the county record.


```python
import re, csv

f = open('1pctsample.txt')
csv_reader = csv.reader(f)
lines = []
for line in csv_reader:
    lines.append(line)
    
len(lines)
```

```python
for line in lines[:10]:
    print line[3]
```

We have addresses and then an occupation.  Occupations may be more than 1 word (see `sales clerk` (two tokens) or `truck drive r` (three tokens).  Right now we want to detect addresses, of which there can be more than one in the row (bad row extraction).

## Writing regex to detect address in Alameda County data

Most addresses seem to follow the pattern (number) (address) (address type).  There are probably other values before and after the address and address type (e.g. 'North' or 'N' can probably show up before and after.  Here's the regex to detect the examples above:

```python
pattern = re.compile('(\d+\w?)\s(?:\w+\s)?(\w+(?:-\w+)?)\s(st|at|ave|blvd|boulevard|road|place|terrace|way)?', re.IGNORECASE)
rows = 0
hits = 0
for line in lines:
    matches = re.findall(pattern, line[3])
    if matches:
        hits += len(matches)
        rows += 1

print "total addresses = " + str(hits)
print "total rows with addresses = " + str(rows)
print "address detection rate is " + str(rows/float(len(lines)))
```

It looks like we can detect the address 99% of the time in each row.  However, there are clearly more addresses detected per row with this regex. Let's see what rows with two addresses detected look like:

```python
for line in lines:
    matches = re.findall(pattern, line[3])
    if len(matches) > 1:
        print line[3]
        break
```

So these are rows that failed to delimit based on party ID.  How many rows have an additional entry in them?

```python
bad_rows = 0
for line in lines:
    matches = re.findall(pattern, line[3])
    if len(matches) > 1:
        bad_rows += 1
print 'total number of rows with more than 1 address = ' + str(bad_rows)
print 'bad row rate is  ' + str(bad_rows/float(len(lines)))
```

It appears about 17% of the rows are bad, in that they have more than one address per line.  We can either throw them out or try to fix them.  For now let's assert that the bad delimiters are distributed at random, and so let's just skip those rows for now.

## Cleaning lines that contain header information: Tabulating precincts

There are lines that contain the header for the page.  We need to be able to detect these and remove the header data.  From scanning the data it seems like the phrase "precinct" (or some variant spelling) reveals that a header has been kept in the data.  Let's explore what we can extract.  We'll write the code using the 1% sample to test, but will run the whole successes data file through the loop in order to get the full distribution of possibilities.

```python
from collections import defaultdict

precincts = defaultdict(int)
header_re = re.compile('(\w+)((?:\w+|\s+|\d+)P\s*R\s*E\s*C\s*I\s*N\s*C\s*T\s*)(?:-)?(\w+(?:\s+))?', re.IGNORECASE)
variant = set(['supplement', 'registration'])
with open('alameda_successes.txt') as infile: # replace the .txt with the 1pct sample file
    for row in infile:
        match = re.search(header_re, row)        
        if match:
            if match.group(1) in variant and match.group(3) is not None:
                precincts[match.group(3).strip().lower()] += 1
            elif match.group(1) is not None:
                val = match.group(1).strip().lower()
                val = filter(lambda x: x.isalpha(), val)
                if len(val) > 0:
                    precincts[val] += 1
                else:
                    precincts['null'] += 1
pprint(dict(precincts))
```

## Improving row delimiting

A loss of 17% of the rows is problematic.  This means that the simple rule of inserting a new line after 'Dem', 'Rep', and 'Declines' is not sufficient.  These are probably OCR transcription errors that appear to be misspellings.  Let's see if we can determine additional delimiters:

```python
from pprint import pprint
bad_rows = 0
transcription_mapping = {'Dem': [r'Dent', r'Dcm', r'Den', r'Dcm'],
                         'Rep': [r'Dent', r'Dcm', r'Den', r'Dcm']}
foo = lines
for line in foo:
    matches = re.findall(pattern, line[3])
    if len(matches) > 1:
        changed = False
        while not changed:
            for sep in transcription_mapping['Dem']:
                if re.search(sep, line[3]):
                    line[3] = re.sub(sep, ','+'Dem'+'\n', line[3])
                    changed = True
            changed = True

for line in foo:
    matches = re.findall(pattern, line[3])
    if len(matches) > 1:
        bad_rows += 1

print 'total number of rows with more than 1 address = ' + str(bad_rows)
print 'bad row rate is ' + str(bad_rows/float(len(foo)))
```
